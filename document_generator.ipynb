```python
import pandas as pd
import math
```

```python
level = 'N3'
file = pd.read_csv('{}.txt'.format(level), sep = '\t')
file.head()
```

Nos quedamos solamente con las 9 primeras columnas (5 primeras definiciones)

```python
file = file[file.columns[0:9]]
```

```python
file.head()
```

Generamos un string que pasará a archivo latex, la plantilla a seguir es:

\card{
     **0:8**
}{ \large{**3**} \\ \Huge{**2**}}

```python
first_part = """
\\documentclass[frontgrid,backgrid,a4paper,11pt]{flacards}
\\usepackage{etoolbox}
\\pretocmd{\\card}{\\def\\curhint{#1}}{}{}

\\pagesetup{3}{7}

\\usepackage{xeCJK}
\\setCJKmainfont{IPAMincho}
\\setmainfont{Linux Libertine O}

\\renewcommand{\\cardtextstylef}{\\small}
\\renewcommand{\\brfoot}{JLPT-N5}


\\renewcommand{\\flhead}{\\footnotesize\\thecardno}
\\renewcommand{\\brhead}{\\footnotesize\\thecardno}

\\begin{document}
"""

last_part = "\end{document}"
```

```python
cards = []
```

En str \\ = \ y \\\\ = \\

```python
s1 = "\card{" + str("DEFINICIONES SEPARADAS POR  ") + "} {\large{" + str("FURIGANA") + "} \\\\ " + "\Huge{" + "KANJI" + "}}" 
print('Haciendo print: \n' + s1 + '\n\nMencionando al final: ')
s1
```

```python
for index, row in file.iterrows():
    # access data using column names
    definitions = row['4'] # De 5 a 8
    for i in range(5, 9):
        if type(row[str(i)]) is str:
            definitions = definitions + ' \\\\ ' + row[str(i)]
    
    # For words in katakana
    furigana = ''
    if type(row['3']) is str:
        furigana = row['3']
    
    si = "\card{" + definitions + "} {\large{" + furigana + "} \\\\ " + "\Huge{" + str(row['2']) + "}}"
    cards.append(si)
```

```python
for i in range(5):
    print(cards[i])
```

```python
f = open("wild_cards/{}_cards.txt".format(level), "a",  encoding="utf-8")
f.write(first_part)

f.write('\n')

for card in cards:
    f.write(card)
    f.write(' \n')

f.write('\n')    

f.write(last_part)
f.close()
```
